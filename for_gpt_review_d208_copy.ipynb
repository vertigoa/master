### **B1. Summarize four assumptions of a multiple linear regression model:**

In multiple linear regression analysis, four key assumptions are critical: linearity between variables, independence of observations, constant error variance (homoscedasticity), and normal distribution of error terms. Understanding and checking these assumptions is essential for the model's reliability and accuracy, providing a solid basis for predictive analytics.

-   **Linearity** asserts that there is a straight-line relationship between each predictor (independent variable) and the response (dependent variable). This means that changes in a predictor variable are associated with proportional changes in the response variable.

-   **Independence of Observations** indicates that the data points in the dataset do not influence each other. Each observation's response is determined by its predictor values, free from the effects of other observations in the dataset.

-   **Homoscedasticity** refers to the requirement that the error terms (differences between observed and predicted values) maintain a consistent variance across all levels of the independent variables. This constant variance ensures that the model's accuracy does not depend on the value of the predictors.

-    **Normality of Errors** involves the assumption that for any fixed value of an independent variable, the error terms are normally distributed. This normal distribution is central to conducting various statistical tests on the model's coefficients to determine their significance.

- **Cardinality**: Considerations for Cardinality in Data Analysis:
Performance: High-cardinality columns can impact the performance of database operations and data analysis tasks. Indexing high-cardinality columns can lead to large indexes that are less efficient.

Data Modeling: Understanding the cardinality of relationships between tables is crucial in database design. Relationships can be one-to-one, one-to-many, or many-to-many, each with different implications for how data is stored and accessed.

Statistical Analysis: Cardinality can affect the choice of statistical methods or machine learning models. For example, high-cardinality categorical variables may need special treatment, such as encoding techniques, before they can be used in models.

Data Visualization: The cardinality of a variable can influence how it should be visualized. High-cardinality variables may be less suitable for certain types of plots (e.g., pie charts) due to the large number of unique values.

Continuous Variables: High cardinality is a natural characteristic of continuous variables because they can take on a wide range of values. In the context of modeling or analysis, the focus might shift from cardinality to the distribution of these values, outliers, or how these variables correlate with the outcome of interest.

Discrete Variables: For discrete variables, high cardinality can present challenges, especially if the variable is categorical. Each unique value of a high-cardinality categorical variable can be considered a separate category, which can lead to issues like sparse data, overfitting, or computational inefficiency in modeling. In these cases, techniques such as binning, feature hashing, or embedding layers in neural networks might be employed to manage high cardinality.

When selecting explanatory variables for creating a multiple linear regression model, especially in the context of analyzing `Initial_days`, several considerations can be made to ensure the model is both accurate and interpretable. These considerations include the relevance of variables to the research question, the type of data and data type, cardinality, multicollinearity, missing values, potential for actionability, statistical and practical significance, and ethical considerations.

Relevance to the Research Question:
Choose variables that are theoretically or empirically known to influence the dependent variable (TotalCharge in this case). This ensures the model is focused on factors that actually matter.

Type of Data and Data Type:
- Categorical Variables:
    - Nominal variables (e.g., Gender, Marital, ReAdmis) should be carefully encoded into numerical values using techniques like one-hot encoding to ensure they can be included in the regression model.
- Continuous Variables:
    - Ratio and interval variables (e.g., Income, VitD_levels, Age) can be directly used in the model but may require scaling or normalization to ensure they contribute appropriately to the model.

Cardinality:
- High-cardinality variables (e.g., Job, City) can introduce a large number of dummy variables if one-hot encoded, which might complicate the model without adding significant predictive power. Consider excluding or aggregating these into broader categories.
- Low-cardinality variables are easier to manage and interpret in the model.

Multicollinearity:
- Avoid variables that are highly correlated with each other, as multicollinearity can distort the importance of explanatory variables and make the model coefficients difficult to interpret.
- Use correlation matrices or variance inflation factor (VIF) analysis to identify and exclude highly correlated variables.

Missing Values:
- Consider the proportion of missing values and the importance of the variable. Variables with a significant amount of missing data might require imputation or exclusion, depending on their relevance and the feasibility of accurately imputing missing values.

Potential for Actionability:
- Prefer variables that, if found to be significant, could lead to actionable insights for hospitals or healthcare providers. This aligns with the project's goal of providing actionable recommendations.

Statistical and Practical Significance:
- Beyond statistical significance, consider the practical significance of variables. A variable might have a statistically significant coefficient but a negligible effect on the dependent variable in practical terms.

Ethical Considerations:
- Ensure that the selection of variables does not inadvertently introduce bias or unethical implications. For example, variables that could lead to discriminatory practices should be handled with caution.

- Detection and treatment of missing values begin by using `df.isnull().sum()`, which returns `True` if the value is null and `False` if the value is not null. `True` values are then summed and returned for each column.
  
- To keep things more efficient, a list of only those variables with missing values will be created by displaying only those columns with `missing values greater than 0`.

```python
# The isnull() method creates a boolean DataFrame indicating which cells in df are null or missing values. 
# The sum() method then sums the number of True values in each column of the boolean DataFrame and returns a 
# new Series with the sums.
null_cols = df.isnull().sum()

#  Create a boolean Series indicating which columns have missing values. 
# Assigns only those values in null_cols that are greater than 0 to missing. (only rows with missing values)
missing = null_cols > 0
if missing.sum() == 0:
    print('No missing values found.')
# Print the count of missing values for each column with missing values, sorted in descending order
print('Number of missing values per column:')
print(null_cols[missing].sort_values(ascending=False))
print(null_cols > 0)
```
